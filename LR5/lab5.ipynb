# Визуализация данных в Python
Этот ноутбук объясняет основы визуализации и демонстрирует часть выполнения заданий.

## Используемые библиотеки

- **numpy** — генерация данных
- **pandas** — работа с таблицами
- **matplotlib** — базовая визуализация
- **seaborn** — расширенная визуализация

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Генерация упрощённых данных о погоде

```python
np.random.seed(42)
n = 1000
df = pd.DataFrame({
    "T": np.random.normal(5, 10, n),
    "U": np.random.uniform(30, 100, n),
    "P": np.random.normal(745, 8, n),
    "N": np.random.randint(0, 101, n),
    "VV": np.random.lognormal(mean=1, sigma=0.5, size=n).clip(0, 20)
})
df.head()
```

## Scatterplot: температура vs влажность (matplotlib)

```python
plt.figure(figsize=(6,4))
plt.scatter(df['T'], df['U'], alpha=0.5)
plt.xlabel('Температура (°C)')
plt.ylabel('Влажность (%)')
plt.title('Scatterplot T vs U')
plt.grid()
plt.show()
```

## Scatterplot с выделением облачности (seaborn)

```python
df['cloud_color'] = np.where(df['N'] == 100, 'blue', 'red')
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='T', y='U', hue='cloud_color', palette=['blue','red'])
plt.title('Scatterplot T vs U with cloudiness color')
plt.show()
```

## Альтернативное задание — генерация X, Y1, Y2

```python
X = np.arange(100)
Y1 = np.random.normal(0,1,100)
Y2 = X + np.random.randint(-2,3,100)
df_alt = pd.DataFrame({'X':X,'Y1':Y1,'Y2':Y2})
df_alt.head()
```

### Scatter X vs Y2

```python
plt.figure(figsize=(6,4))
plt.scatter(df_alt['X'], df_alt['Y2'])
plt.xlabel('X')
plt.ylabel('Y2')
plt.title('Scatter X vs Y2')
plt.grid()
plt.show()
```

### Линейный график зависимости Y1 от X

```python
plt.figure(figsize=(6,4))
plt.plot(df_alt['X'], df_alt['Y1'])
plt.xlabel('X')
plt.ylabel('Y1')
plt.title('Line plot: Y1 by X')
plt.grid()
plt.show()
```
